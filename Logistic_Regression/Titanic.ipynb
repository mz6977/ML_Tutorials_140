```python
import pandas as pd
```

```python
# loading the dataset
df_train = pd.read_csv('/kaggle/input/titanic/train_data.csv')
df_train
```

```python
# dropping unnecessary columns(first 2)
df_train.drop(['Unnamed: 0', 'PassengerId'], axis=1,inplace = True)
df_train
```

```python
# taking Survived column in Y_train and other columns in X_train
X_train = df_train.iloc[:,1:]
Y_train = df_train.iloc[:,:1]
```

```python
X_train
```

```python
# chceking is there any missing values
for column in X_train.columns:
    print(X_train[column].isna().value_counts())
```

```python
features = X_train.columns
target = Y_train.columns
```

```python
import numpy as np
```

```python
# performing Correlation analysis to find unnecessary columns 
significant_features = []
threshold = 0.5
for col in X_train:
    corelation = np.corrcoef(X_train[col],Y_train['Survived'])[0,1]
    if(corelation >= threshold):
        significant_features.append(col)
        
significant_features
```

```python
# importing Logistic Regression from sklearn library
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# training the model with train data
model.fit(X_train,Y_train)
```

```python
# loading test dataset
df_test = pd.read_csv('/kaggle/input/titanic/test_data.csv')
df_test
```

```python
df_test.drop(['Unnamed: 0', 'PassengerId'], axis=1,inplace = True)
df_test
```

```python
X_test = df_test.iloc[:,1:]
Y_test = df_test.iloc[:,:1]
```

```python
#traing model with test data
Y_predicted = model.predict(X_test)
```

```python
Y_test
```

```python
# renameing Survived column to Y_true
Y_test.rename(columns = {'Survived':'Y_True'},inplace = True)
```

```python
# creating new dataframe with the predicted result
Y_predicted = pd.DataFrame({'Y_predicted': Y_predicted})
```

```python
# merging two df
result_df = pd.merge(Y_test,Y_predicted, left_index=True, right_index=True)
result_df.head(50)
```

```python
# create new column to easily visualize whether they matched or not
result_df['Match'] = np.where(result_df['Y_True'] == result_df['Y_predicted'], 'Yes', 'No')
result_df
```

```python
result_df['Match'].value_counts()
```

```python
model.score(X_test,Y_test)
```
